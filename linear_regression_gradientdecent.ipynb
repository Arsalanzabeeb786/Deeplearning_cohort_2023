```python

# Sample data for demonstration
X = [1, 2, 3, 4, 5]
Y = [2, 4, 5, 4, 5]

w = 0
b = 0

L = 0.001  # The learning Rate
epochs = 1500 # The number of iterations to perform gradient descent

m = float(len(X)) # Number of elements in X
start_time = time.time()
# Performing Gradient Descent
for i in range(epochs):
    Y_pred = sum(w * x + b for x in X)  # The current predicted value of Y
    j = sum(map(lambda x : x**2 ,list((Y_pred-y) for y in Y)))
    print(j)
    D_w = sum(x * (Y_pred - y) for x, y in zip(X, Y))  # Derivative wrt w
    D_w = D_w / m
    D_b = sum(Y_pred - y for y in Y)  # Derivative wrt b
    D_b = D_b / m

    # Check for NaN or inf values in derivatives
    if D_w != D_w or D_b != D_b or D_w == float('inf') or D_b == float('inf'):
        print("NaN or inf value encountered. Adjust learning rate.")
        break

    w = w - (L * D_w)  # Update w
    b = b - (L * D_b)  # Update b
end_time = time.time()
# Calculate and print the time taken
elapsed_time = end_time - start_time
print("Time taken:", elapsed_time, "seconds")
print(w, b)

# Plotting the data points and regression line
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, [w * x + b for x in X], color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

# lets use library to build same

```python
from sklearn.linear_model import LinearRegression

# Sample data for demonstration
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 5, 4, 5])

# Create a LinearRegression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Get the coefficients and intercept
w = model.coef_[0]
b = model.intercept_

print("Final values:", w, b)


# Plotting the data points and regression line
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
